220 resumes annotated manually with 11 entity types.

````python
import json
import re
import random
from pathlib import Path
from collections import Counter

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

colors = ['#4361ee', '#3a86ff', '#ff006e', '#fb5607', '#8338ec',
          '#06d6a0', '#118ab2', '#ffd166', '#ef476f', '#073b4c']
````

## 1. Raw Data Structure

The file is **JSONL** — each line is one resume as a JSON object.

```json
{
  "content": "Abhishek Jha\nApplication Development Associate...",
  "annotation": [
    {
      "label": ["Skills"],
      "points": [{"start": 993, "end": 1153, "text": "C, Java..."}]
    }
  ]
}
```

Entity labels in the dataset: `Name`, `College Name`, `Degree`, `Graduation Year`,  
`Years of Experience`, `Companies worked at`, `Designation`, `Skills`,  
`Location`, `Email Address`, `Links`

````python
data_path = Path('../data/raw/Entity Recognition in Resumes.json')

raw_data = []
with open(data_path, encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if line:
            raw_data.append(json.loads(line))

print(f'Total resumes loaded: {len(raw_data)}')
print()

# show what one resume looks like
sample = raw_data[0]
print('First 300 characters of resume content:')
print(sample['content'][:300])
````

## 2. Problems Found in the Raw Data

Before training we need to understand what is wrong with the raw annotations.

| Problem | Count | Why it matters |
|---------|-------|----------------|
| Inclusive `end` index | all spans | spaCy requires exclusive end → needs `end + 1` |
| Overlapping spans | 117 pairs | spaCy raises `ValueError` during training |
| `\r\n` / `\xa0` in text | many records | shifts character offsets → wrong span extraction |
| Inconsistent label casing | e.g. `Skills` vs `SKILLS` | NER trainer needs consistent label names |
| Very short resumes (< 50 words) | 15 records | likely parsing errors, not real CVs |

````python
# count words per resume
word_counts = [len(r['content'].split()) for r in raw_data]

# count how many times each label appears
label_counts = Counter()
for r in raw_data:
    for ann in (r.get('annotation') or []):
        for lbl in (ann.get('label') or []):
            label_counts[lbl.strip()] += 1

# count overlapping spans and offset mismatches
total_overlaps = 0
total_offset_errors = 0

for r in raw_data:
    text = r.get('content', '')
    spans = []
    for ann in (r.get('annotation') or []):
        for pt in (ann.get('points') or []):
            s, e = pt['start'], pt['end']
            # check if the annotated text matches what is actually at [s:e+1]
            if text[s:e+1].strip() != pt.get('text', '').strip():
                total_offset_errors += 1
            spans.append((s, e))
    # check every pair of spans for overlap
    for i in range(len(spans)):
        for j in range(i + 1, len(spans)):
            s1, e1 = spans[i]
            s2, e2 = spans[j]
            if s1 <= e2 and s2 <= e1:
                total_overlaps += 1

print('Data quality summary:')
print(f'  Total resumes           : {len(raw_data)}')
print(f'  Too short (< 50 words)  : {sum(1 for w in word_counts if w < 50)}  → will be dropped')
print(f'  Overlapping span pairs  : {total_overlaps}  → crashes spaCy training')
print(f'  Offset mismatches       : {total_offset_errors}  → caused by \\r\\n line endings')
print(f'  Average words / resume  : {int(sum(word_counts) / len(word_counts))}')
````

````python
fig, axes = plt.subplots(1, 2, figsize=(13, 4))
fig.suptitle('Raw Dataset — Overview', fontweight='bold', fontsize=13)

# word count distribution
axes[0].hist(word_counts, bins=25, color=colors[0], edgecolor='white')
axes[0].axvline(50, color=colors[2], linestyle='--',
                label='Drop threshold (50 words)')
axes[0].axvline(int(sum(word_counts) / len(word_counts)), color=colors[1],
                linestyle='-', label=f'Mean ({int(sum(word_counts)/len(word_counts))} words)')
axes[0].set_title('Word Count Distribution')
axes[0].set_xlabel('Word count')
axes[0].set_ylabel('Number of resumes')
axes[0].legend()
axes[0].grid(alpha=0.3)

# label distribution
lbls, cnts = zip(*label_counts.most_common())
bars = axes[1].barh(lbls, cnts, color=colors[:len(lbls)], edgecolor='white')
axes[1].bar_label(bars, padding=3, fontsize=8)
axes[1].set_title('Raw Label Distribution')
axes[1].invert_yaxis()
axes[1].grid(alpha=0.3, axis='x')

plt.tight_layout()
plt.show()
````

## 3. Cleaning Steps

| Step | What we do | Why |
|------|-----------|-----|
| Fix end index | `end = end + 1` | Dataturks uses inclusive end; spaCy needs exclusive |
| Normalize text | Remove `\r\n`, `\xa0`, HTML entities | These shift character offsets and break span extraction |
| **Trim span whitespace** | Shrink boundaries to actual content | **spaCy raises E024 if a span has leading/trailing whitespace** |
| Skip punct-start spans | Drop spans starting with punctuation | spaCy E024 — punct at span boundaries causes training crash |
| Remove overlapping spans | Keep the longer span | spaCy raises `ValueError` on overlapping entities |
| Normalize label names | Map to `UPPER_SNAKE_CASE` | Consistent naming required for NER training |
| Drop short resumes | Remove if < 50 words | These are likely file parsing errors, not real CVs |


````python
import string

# maps raw annotator labels to a consistent canonical format
LABEL_MAP = {
    'name':                'NAME',
    'college name':        'COLLEGE_NAME',
    'degree':              'DEGREE',
    'graduation year':     'GRADUATION_YEAR',
    'years of experience': 'YEARS_OF_EXPERIENCE',
    'companies worked at': 'COMPANIES_WORKED_AT',
    'designation':         'DESIGNATION',
    'skills':              'SKILLS',
    'location':            'LOCATION',
    'email address':       'EMAIL_ADDRESS',
    'links':               'LINKS',
    'unknown':             None,   # 2 occurrences in the dataset — we drop these
}


def clean_text(text):
    # replace Windows-style line endings — they add an extra byte per line
    # which shifts every character offset after them
    text = text.replace('\r\n', '\n').replace('\r', '\n')

    # non-breaking space looks invisible but breaks offset calculations
    text = text.replace('\xa0', ' ')

    # some resumes were copy-pasted from web pages and carry HTML entities
    text = text.replace('&amp;', '&').replace('&lt;', '<').replace('&gt;', '>')

    # collapse 3+ consecutive blank lines into 2
    text = re.sub(r'\n{3,}', '\n\n', text)

    return text.strip()


def extract_spans(annotation, text):
    spans = []
    for ann in (annotation or []):
        labels = ann.get('label', [])
        if isinstance(labels, str):
            labels = [labels]
        for pt in ann.get('points', []):
            start = pt['start']
            end   = pt['end'] + 1  # convert inclusive end → exclusive end for spaCy

            # clamp to valid range in case of annotation errors
            start = max(0, min(start, len(text)))
            end   = max(0, min(end,   len(text)))

            if start >= end:
                continue

            span_text = text[start:end]

            # spaCy raises E024 if a span has leading or trailing whitespace
            # so we shrink the boundaries inward to the actual content
            stripped = span_text.strip()
            if not stripped:
                continue

            leading  = len(span_text) - len(span_text.lstrip())
            trailing = len(span_text) - len(span_text.rstrip())
            start    = start + leading
            end      = end - trailing if trailing > 0 else end

            if start >= end:
                continue

            # spaCy also rejects spans that start with punctuation
            if text[start] in string.punctuation:
                continue

            for lbl in labels:
                mapped = LABEL_MAP.get(lbl.strip().lower())
                if mapped:
                    spans.append((start, end, mapped))
    return spans


def remove_overlaps(spans):
    # sort by span length descending so longer spans get priority
    sorted_spans = sorted(spans, key=lambda s: s[1] - s[0], reverse=True)
    kept = []
    for span in sorted_spans:
        s, e, _ = span
        # only keep this span if it does not overlap with any already accepted span
        has_overlap = any(s < ek and e > sk for sk, ek, _ in kept)
        if not has_overlap:
            kept.append(span)
    # return spans in document order (left to right)
    return sorted(kept, key=lambda s: s[0])
````

````python
processed = []
dropped   = 0

for record in raw_data:
    text = clean_text(record.get('content', ''))

    # drop resumes that are too short to be real CVs
    if len(text.split()) < 50:
        dropped += 1
        continue

    spans = extract_spans(record.get('annotation'), text)
    spans = remove_overlaps(spans)

    processed.append({
        'text':     text,
        'entities': [[s, e, lbl] for s, e, lbl in spans]
    })

print(f'Valid records   : {len(processed)}')
print(f'Dropped records : {dropped}  (< 50 words)')
````

````python
# use the same 200/20 split ratio as the original dataset paper
random.seed(42)
random.shuffle(processed)

split_index = int(len(processed) * 200 / 220)
train_data  = processed[:split_index]
test_data   = processed[split_index:]

print(f'Train : {len(train_data)} resumes')
print(f'Test  : {len(test_data)} resumes')
````

````python
output_dir = Path('../data/processed')
output_dir.mkdir(exist_ok=True)

with open(output_dir / 'train_data.json', 'w', encoding='utf-8') as f:
    json.dump(train_data, f, ensure_ascii=False)

with open(output_dir / 'test_data.json', 'w', encoding='utf-8') as f:
    json.dump(test_data, f, ensure_ascii=False)

print('Saved: train_data.json and test_data.json → data/processed/')
````

````python
# count entity labels in the cleaned data
label_after = Counter()
for item in processed:
    for _, _, lbl in item['entities']:
        label_after[lbl] += 1

# map original raw label names to canonical names for a fair comparison
label_before_mapped = {}
for lbl, count in label_counts.items():
    canonical = LABEL_MAP.get(lbl.lower())
    if canonical:
        label_before_mapped[canonical] = label_before_mapped.get(canonical, 0) + count

all_labels  = sorted(label_after.keys())
x = np.arange(len(all_labels))
w = 0.35

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Before vs After Preprocessing', fontweight='bold', fontsize=13)

before_vals = [label_before_mapped.get(l, 0) for l in all_labels]
after_vals  = [label_after[l]               for l in all_labels]

b1 = axes[0].bar(x - w/2, before_vals, w, label='Before', color=colors[2], alpha=0.8, edgecolor='white')
b2 = axes[0].bar(x + w/2, after_vals,  w, label='After',  color=colors[1], alpha=0.8, edgecolor='white')
axes[0].bar_label(b1, fontsize=7, padding=2)
axes[0].bar_label(b2, fontsize=7, padding=2)
axes[0].set_xticks(x)
axes[0].set_xticklabels(all_labels, rotation=40, ha='right', fontsize=8)
axes[0].set_ylabel('Number of annotations')
axes[0].set_title('Label Distribution Before vs After Cleaning')
axes[0].legend()
axes[0].grid(alpha=0.3, axis='y')

axes[1].pie(
    [len(train_data), len(test_data)],
    labels=[f'Train\n{len(train_data)}', f'Test\n{len(test_data)}'],
    colors=[colors[0], colors[1]],
    autopct='%1.0f%%',
    wedgeprops={'edgecolor': 'white', 'width': 0.6},
    startangle=90
)
axes[1].set_title('Train / Test Split')

plt.tight_layout()
plt.show()
````
